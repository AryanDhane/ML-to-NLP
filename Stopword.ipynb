```python
paragraph ="""Mahatma Gandhi was born on 2nd October 1869 in Porbandar, Gujarat. His full name was Mohandas Karamchand Gandhi. He was a great leader, freedom fighter, and a symbol of truth and non-violence.

Gandhiji believed in Ahimsa (non-violence) and Satya (truth). He showed the world that even the strongest power can be defeated without violence. Through movements like the Non-Cooperation Movement, Salt March, and Quit India Movement, he played a key role in India’s struggle for independence.

He lived a very simple life and always worked for the poor, the weak, and the untouchables. Gandhiji taught us important values such as honesty, patience, self-discipline, and peace.

Because of his great principles and leadership, India got freedom on 15th August 1947. His life and teachings still inspire millions of people across the world.

Let us all follow his path of truth and non-violence to make our society and nation better viswasan."""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
sentences = nltk.sent_tokenize(paragraph)
type(sentences)
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [ps.stem(word) for word in words if not word in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
#  applay stopword and filter and then apply snowball stemmer
from nltk.stem import SnowballStemmer
ps = SnowballStemmer(language='english')
sentences = nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [ps.stem(word) for word in words if not word in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```
