```python
sentance = """Mahatma Gandhi, also known as the Father of the Nation in India, was a great leader and freedom fighter. He was born on October 2, 1869, in Porbandar, Gujarat. Gandhi's philosophy of non-violence and truth inspired millions of people around the world."""
```

```python
import nltk
words = nltk.word_tokenize(sentance)
pt=nltk.pos_tag(words)
```

```python
ne_chunk = nltk.ne_chunk(pt).draw()
```
