# Stemming

## porter stemmer

```python
words = ["playing", "played", "plays", "studies"]
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

for word in words:
    print(word+ "â†’"+ ps.stem(word))
```

```python
ps.stem("sitting")
```

## Regular expences stemmer

```python
from nltk.stem import RegexpStemmer

rs = RegexpStemmer('ing$|s$|e$|able$', min=4)

for word in words:
    print(word+ "â†’"+ rs.stem(word)) 
```

# Snowball Stemming

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

words = ["playing", "played", "plays", "studies"]

for w in words:
    print(w, "â†’", stemmer.stem(w))
```
